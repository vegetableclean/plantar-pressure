```python
def collect_location(a,b):
    x=r'H:\experiment\generation\validation\aug_Left_foot\%s'%a
    y=r'H:\experiment\generation\validation\aug_Right_foot\%s'%b
    return x,y
```

```python
def random_image(fileDir):   
    pathDir = os.listdir('%s'%fileDir)
    sample = random.sample(pathDir, 20)  #隨機找三張
    return sample
```

```python
def read_file(a,b,e,f):
    match=[]
    match.append(r'H:\experiment\generation\validation\aug_Left_foot\%s\%s'%(a,e))
    match.append(r'H:\experiment\generation\validation\aug_Right_foot\%s\%s'%(b,f))
    print('now we are combine from left %s and right %s to get %s and %s'%(a,b,e,f))
    return match
```

```python
def combine_photo(a,b,e,f,arr):
    #arr是一個列表，裡面是兩個lis圖片的路徑，例如["p1.png", 'p2.png']
    toImage = Image.new('RGB', (38, 35))
    img1 = Image.open(arr[0])
    img12 = Image.open(arr[1])
    toImage.paste(img1, (3, 3))
    toImage.paste(img12, (25, 3, 25+ img1.size[0] , 3+ img1.size[1]))
    #函式描述：toImage:背景圖片,paste()函式四個變數分別為：起始橫軸座標，起始縱軸座標，橫軸結束座標，縱軸結束座標；
    toImage.save(r"H:\experiment\generation\validation\output\%s%s\%s_%s.jpg"%(a,b,e,f),"JPEG")
    plt.imshow(toImage)
    plt.title('plantar')
    #plt.show()
```
